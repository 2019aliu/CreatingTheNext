```python
def calculate_poor(column, threshold):
    total = 0
    for i in SE_data[column][0:threshold]:
        total += i
    return total
```

```python
asian_poor = calculate_poor('count_asian_pooled', 15)
black_poor = calculate_poor('count_black_pooled', 15)
hisp_poor = calculate_poor('count_hisp_pooled', 15)
white_poor = calculate_poor('count_white_pooled', 15)
```

```python
def print_year_unemployment_values(collection, column, year):
    values = []
    for i in range(len(collection['date'])):
        if (collection['date'][i].startswith(year)):
            values.append(collection[column][i])
    return values

UA2019 = print_year_unemployment_values(UA, 'unemployment_asian', '2019')
UB2019 = print_year_unemployment_values(UB, 'unemployment_black', '2019')
UL2019 = print_year_unemployment_values(UL, 'unemployment_latino', '2019')
UW2019 = print_year_unemployment_values(UW, 'unemployment_white', '2019')
```

```python
# Poor / Unemployment Average in 2019

asian_prop = asian_poor / (sum(UA2019) / len(UA2019))
black_prop = black_poor / (sum(UB2019) / len(UB2019))
hisp_prop = hisp_poor / (sum(UL2019) / len(UL2019))
white_prop = white_poor / (sum(UW2019) / len(UW2019))
```

```python
print(asian_prop)
print(black_prop)
print(hisp_prop)
print(white_prop)
```
